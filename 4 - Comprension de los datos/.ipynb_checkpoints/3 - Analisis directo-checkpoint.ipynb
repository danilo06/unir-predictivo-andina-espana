### **Notebook 4: Etapa de comprension de los datos - Analisis especifico**  


```python
import seaborn as sns
import matplotlib.pyplot as plt
from google.cloud import bigquery # Cliente de GCP
```

```python
client = bigquery.Client()
table_id = "unir-predictiv0-andina-espana.datacomex.comex_comunidad_andina"
```

```python
def create_chart(column_name, df):
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    axes[0].boxplot(df[column_name], vert=False, patch_artist=True, 
                    boxprops=dict(facecolor='royalblue'))
    axes[0].set_title("Diagrama de Bigotes")
    axes[0].set_xlabel(column_name)

    sns.violinplot(x=column_name, data=df, ax=axes[1], color='royalblue')
    axes[1].set_title("Violin Plot - Distribución de Datos")
    axes[1].set_xlabel(column_name)
    
    axes[2].hist(df[column_name], bins=50, color='royalblue', edgecolor='black')
    axes[2].set_title("Histograma de Datos")
    axes[2].set_xlabel(column_name)
    axes[2].set_ylabel("Frecuencia")
    
    plt.tight_layout()
    plt.show()

def get_bigquery_data(sql):
    query_job = client.query(sql)
    return query_job.to_dataframe()
    
```

### Explorando Euros -- Q3 (20174.22, tercer cuartil - 75%)

```python
query = f"SELECT euros FROM `{table_id}` where euros < 20174.22"
df = get_bigquery_data(query)
create_chart('euros', df)
```

### Explorando Dolares -- Q3 (23823.99, tercer cuartil - 75%)

```python
query = f"SELECT dolares FROM `{table_id}` where dolares < 23823.99"
df = get_bigquery_data(query)
create_chart('dolares', df)
```

### Explorando kilogramos -- Q3 (3150, tercer cuartil - 75%)

```python
query = f"SELECT kilogramos FROM `{table_id}` WHERE kilogramos < 3150"
df = get_bigquery_data(query)
create_chart('kilogramos', df)
```
