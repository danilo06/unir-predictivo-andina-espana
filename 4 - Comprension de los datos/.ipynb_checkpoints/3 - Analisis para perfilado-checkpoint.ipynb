### **Notebook 4: Etapa de comprension de los datos - Analisis para perfilado**  


```python
import seaborn as sns
import matplotlib.pyplot as plt
from google.cloud import bigquery
```

```python
client = bigquery.Client()
table_id = "unir-predictiv0-andina-espana.datacomex.comex_comunidad_andina"
```

```python
def create_chart(column_name, dfs, levels):
    for i, df in enumerate(dfs):
        fig, axes = plt.subplots(1, 3, figsize=(9, 3))

        axes[0].boxplot(df[column_name], vert=False, patch_artist=True, 
                        boxprops=dict(facecolor='royalblue'))
        axes[0].set_title(f"Diagrama de Bigotes - Nivel {levels[i]}")
        axes[0].set_xlabel(column_name)

        sns.violinplot(x=column_name, data=df, ax=axes[1], color='royalblue')
        axes[1].set_title(f"Violin Plot - Distribución de Datos - Nivel {levels[i]}")
        axes[1].set_xlabel(column_name)

        axes[2].hist(df[column_name], bins=50, color='royalblue', edgecolor='black', alpha=0.5)
        axes[2].set_title(f"Histograma de Datos - Nivel {levels[i]}")
        axes[2].set_xlabel(column_name)
        axes[2].set_ylabel("Frecuencia")
        
        plt.tight_layout()
        plt.show()

def get_bigquery_data(sql):
    query_job = client.query(sql)
    return query_job.to_dataframe()
```

### Comprendiendo relacion entre Euros Y Dolares

```python

query = f"SELECT euros, dolares FROM `{table_id}`"
df = get_bigquery_data(query)

correlation_matrix = df[['euros', 'dolares']].corr()
print(correlation_matrix)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", ax=ax[0])
ax[0].set_title("Mapa de Correlación")
ax[1].scatter(df['euros'], df['dolares'])
ax[1].set_title("Relación entre euros y dolares")
ax[1].set_xlabel("euros")
ax[1].set_ylabel("dolares")
plt.tight_layout()
plt.show()
```

### Relacion entre dinero y kilogramos

```python
query = f"SELECT kilogramos, dolares FROM `{table_id}`"
df = get_bigquery_data(query)

correlation_matrix = df[['dolares', 'kilogramos']].corr()
print(correlation_matrix)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", ax=ax[0])
ax[0].set_title("Mapa de Correlación")
ax[1].scatter(df['dolares'], df['kilogramos'])
ax[1].set_title("Relación entre dolares y kilogramos")
ax[1].set_xlabel("dolares")
ax[1].set_ylabel("kilogramos")
plt.tight_layout()
plt.show()
```

### Explorando Dolares -- Q3 (tercer cuartil - 75%)

```python
levels = [
    {"level": 1, "Q3": 62114.445},
    {"level": 2, "Q3": 30263.09},
    {"level": 3, "Q3": 20755.97},
    {"level": 4, "Q3": 19117.312},
    {"level": 5, "Q3": 18907.998}
]

dfs = []
for level_info in levels:
    nivel = level_info["level"]
    q3_value = level_info["Q3"]
    query = f"SELECT dolares FROM `{table_id}` WHERE comex_nivel_taric = '{nivel}' AND dolares < {q3_value}"
    print(query)
    df = get_bigquery_data(query)
    dfs.append(df)

create_chart('dolares', dfs, [level_info["level"] for level_info in levels])
```

### Explorando kilogramos -- Q3 (tercer cuartil - 75%)

```python
levels = [
    {"level": 1, "Q3": 16048},
    {"level": 2, "Q3": 4418},
    {"level": 3, "Q3": 2402},
    {"level": 4, "Q3": 2200},
    {"level": 5, "Q3": 2170}
]

dfs = []
for level_info in levels:
    nivel = level_info["level"]
    q3_value = level_info["Q3"]
    query = f"SELECT kilogramos FROM `{table_id}` WHERE comex_nivel_taric = '{nivel}' AND kilogramos < {q3_value}"
    print(query)
    df = get_bigquery_data(query)
    dfs.append(df)

create_chart('kilogramos', dfs, [level_info["level"] for level_info in levels])
```
